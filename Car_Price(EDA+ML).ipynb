## KNN


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ad_data = pd.read_csv("Social_Network_Ads.csv")
ad_data
```

```python
# exploratory data analysis

sns.pairplot(ad_data, hue= 'Purchased')
```

```python
ad_data["Gender_band"] = 0
ad_data.loc[ad_data["Gender"]== "Male", "Gender_band"] = 0
ad_data.loc[ad_data["Gender"]== "Female", "Gender_band"] = 1
ad_data_cleaned = ad_data.drop(["User ID", "Gender"], axis=1)
ad_data_cleaned
```

```python
sns.pairplot(ad_data_cleaned, hue= 'Purchased')
```

```python
ad_data_cleaned.describe()
```

```python
# import StandardScaler class
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit( ad_data_cleaned.drop(["Purchased", "Gender_band"], axis=1)   )
```

```python
# preform the scaling transformation

scaled_features = scaler.transform(ad_data_cleaned.drop(["Purchased", "Gender_band"], axis=1))
scaled_features
```

```python
df = pd.DataFrame(scaled_features, columns=ad_data_cleaned.columns[0:2])
df
```

```python
df = df.join(ad_data_cleaned[['Purchased', 'Gender_band']])
df
```

```python
X = df[['Gender_band', 'EstimatedSalary', 'Age']]
y = df['Purchased']

X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# training

knn.fit(X_train, y_train)
```

```python
# Prediction

prediction = knn.predict(X_test)
prediction
```

```python
np.array(y_test)
```

```python
# evaluation

from sklearn.metrics import confusion_matrix, accuracy_score

print(accuracy_score(y_test, prediction))
```

```python
print(confusion_matrix(y_test, prediction))
```
